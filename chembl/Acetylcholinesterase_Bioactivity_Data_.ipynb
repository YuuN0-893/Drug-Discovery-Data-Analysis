```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
protein = new_client.target
query= protein.filter(pref_name__iexact="Acetylcholinesterase")
targets=pd.DataFrame.from_dict(query)
targets
```

```python
s_target= targets.target_chembl_id[0]
s_target
```

```python
activity= new_client.activity
res= activity.filter(target_chembl_id=s_target).filter(standard_type="IC50")
```

```python
df=pd.DataFrame.from_dict(res)
```

```python
df.head()
```

```python
df.to_csv('acetylcholinesterase_01_bioactivity_data_raw.csv', index=False)
```
